```python
from General_Fetch import Fetch
from obspy import UTCDateTime
```

```python
network = None
station = None
starttime = '2010-01-01'
endtime = '2018-01-01'

minlat = 16
minlon = 105
maxlat = 50
maxlon = 150
```

```python
test = Fetch(network=network,station=station,starttime=UTCDateTime(starttime),endtime=UTCDateTime(endtime),
minlatitude=minlat,maxlatitude=maxlat,minlongitude=minlon,maxlongitude=maxlon)
```

```python
#Get all the station information
test.fetchInventory()
```

```python
#get the inventory and plot it 
import matplotlib.pyplot as plt
%matplotlib inline

inv = test.inventory
figure = plt.figure(figsize=(20,20))
fig = inv.plot(method='basemap',projection='local',color_per_network='True',size=10,show=False)
plt.show()
```

```python
for network in inv:
    
    print('Network %s has %s stations' %(network.code,len(network)))
```
